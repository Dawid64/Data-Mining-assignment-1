```python
from sklearn.datasets import load_iris
import simpsom as sps
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris_data=iris.data
iris_target=iris.target
```

```python
net=sps.SOMNet(20,20,iris_data,PBC=True,GPU=False)
net.train(epochs=100)
```

```python
net.project(iris_data,labels=iris_target,show=True)
```

```python
net.cluster(iris_data,show=True)
```

```python
spaceship=pd.read_csv('../../spaceship-titanic/train.csv')
```

```python
spaceship.dropna(inplace=True)
spaceship.drop(['PassengerId','HomePlanet','CryoSleep','Cabin','Destination','VIP','Name'],axis=1,inplace=True)
spaceship_data=spaceship.iloc[:,:-1]
spaceship_data=spaceship_data.to_numpy()
```

```python
spaceship_target=spaceship.iloc[:,-1]
spaceship_target=pd.DataFrame(0 if i == False else 1 for i in spaceship_target.values)
spaceship_target=spaceship_target.to_numpy()
```

```python
net=sps.SOMNet(20,20,spaceship_data,PBC=True,GPU=False)
net.train(epochs=100)
```

```python
net.project(spaceship_data,labels=spaceship_target,show=True)
```

```python
net.cluster(spaceship_data,show=True)
```

```python
from sklearn.datasets import load_iris
import simpsom as sps
```

```python
iris = load_iris()
```

```python
iris_data=iris.data
iris_target=iris.target
```

```python
net=sps.SOMNet(20,20,iris_data,PBC=True,GPU=False)
net.train(epochs=100)
net.plot_convergence()
```

```python
net.diff_graph()
```

```python
net.nodes_graph()
```

```python
net.project(iris_data,labels=iris_target,show=True,print_out=True)
```

```python
net.cluster(iris_data)
```

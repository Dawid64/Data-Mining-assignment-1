```python
import pandas as pd
from sklearn_som.som import SOM
from sklearn.preprocessing import MinMaxScaler
spaceship=pd.read_csv('spaceship-titanic/train.csv')
spaceship
```

```python
spaceship.dropna(inplace=True)
spaceship.drop(['PassengerId','HomePlanet','CryoSleep','Cabin','Destination','VIP','Name','Age'],axis=1,inplace=True)
spaceship_data=spaceship.iloc[:,:-1]
norm=MinMaxScaler(feature_range=(0,5))
norm.fit(spaceship_data)
spaceship_data=pd.DataFrame(norm.transform(spaceship_data),columns=spaceship_data.columns)
spaceship_data
```

```python
spaceship_target=spaceship.iloc[:,-1]
spaceship_target=pd.DataFrame(0 if i == False else 1 for i in spaceship_target.values)
spaceship_target
```

```python
spaceship_som=SOM(m=3,n=2,dim=spaceship_data.shape[1])
```

```python
spaceship_som.fit(spaceship_data)
spaceship_som.predict(spaceship_data)
```
